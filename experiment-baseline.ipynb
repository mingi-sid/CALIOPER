```python
jigsaw_features = ['jigsaw_toxicity', 'jigsaw_severe_toxicity', 'jigsaw_identity_attack',
       'jigsaw_insult', 'jigsaw_profanity', 'jigsaw_threat',
       'jigsaw_sexually_explicit', 'jigsaw_flirtation',
       'jigsaw_attack_on_author', 'jigsaw_attack_on_commenter',
       'jigsaw_incoherent', 'jigsaw_inflammatory', 'jigsaw_likely_to_reject',
       'jigsaw_obscene', 'jigsaw_spam', 'jigsaw_unsubstantial']
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
```

```python
# Find the best feature to predict offensive label
best_feature = ''
best_f1 = 0
best_threshold = 0
for feature in jigsaw_features:
    thresholds = [0.01 * i for i in range(1, 100)]
    for threshold in thresholds:
        y_pred = train_data[feature].apply(lambda x: 'Y' if x >= threshold else 'N')
        f1_score_this = f1_score(train_data['offensive'], y_pred, pos_label='Y')
        if f1_score_this > best_f1:
            best_f1 = f1_score_this
            best_feature = feature
            best_threshold = threshold
print(f"Best feature: {best_feature}, best F1 score: {best_f1}, threshold: {best_threshold}")
```

```python
# Implement Charater-level CNN using pytorch
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np

# Define a class for the dataset
class OffensiveDataset(Dataset):
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels
        self.vocab = set()
        for text in self.data:
            self.vocab.update(text)
        self.vocab = sorted(list(self.vocab))
        self.char2idx = {c: i for i, c in enumerate(self.vocab)}
        self.idx2char = {i: c for i, c in enumerate(self.vocab)}
        
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        # padding data, maximum length = 100
        text = self.data[idx]
        if len(text) > 100:
            text = text[:100]
        else:
            text = text + ' ' * (100 - len(text))
        label = self.labels[idx]
        x = torch.zeros(len(text), dtype=torch.int)
        for i, c in enumerate(text):
            x[i] = self.char2idx[c]
        return x, label
    
    def get_vocab_size(self):
        return len(self.vocab)

# Create the dataset from dataframe, ignore index
# (train_data['offensive'] == 'Y') to torch Tensor
train_dataset = OffensiveDataset(train_data['text'].tolist(), (train_data['offensive'] == 'Y').astype(int).tolist())
test_dataset = OffensiveDataset(test_data['text'].tolist(), (test_data['offensive'] == 'Y').astype(int).tolist())

# Create the dataloaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Define the model
class CharCNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, num_filters, filter_sizes, output_dim, dropout):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.convs = nn.ModuleList([
            nn.Conv2d(in_channels=1, out_channels=num_filters, kernel_size=(fs, embedding_dim)) for fs in filter_sizes
        ])
        self.fc = nn.Linear(len(filter_sizes) * num_filters, output_dim)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, x):
        # x = [sent len, batch size]
        embedded = self.embedding(x)
        # embedded = [sent len, batch size, emb dim]
        embedded = embedded.unsqueeze(1)
        # embedded = [1, batch size, sent len, emb dim]
        conved = [F.relu(conv(embedded)).squeeze(3) for conv in self.convs]
        # conv_n = [batch size, num_filters, sent len - filter_sizes[n] + 1]
        pooled = [F.max_pool1d(conv, conv.shape[2]).squeeze(2) for conv in conved]
        # pooled_n = [batch size, num_filters]
        cat = self.dropout(torch.cat(pooled, dim=1))
        # cat = [batch size, num_filters * len(filter_sizes)]
        # output = [batch size, output_dim]
        return self.fc(cat)
    
# Define the hyperparameters
vocab_size = train_dataset.get_vocab_size()
embedding_dim = 100
num_filters = 100
filter_sizes = [2, 3, 4]
output_dim = 1
dropout = 0.5
epochs = 50

# Initialize the model
model = CharCNN(vocab_size, embedding_dim, num_filters, filter_sizes, output_dim, dropout)

# Define the optimizer and loss function
optimizer = optim.Adam(model.parameters())
criterion = nn.BCEWithLogitsLoss()

# Define the device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Move the model to the device
model = model.to(device)
criterion = criterion.to(device)

# Define the training function
def train(model, iterator, optimizer, criterion):
    epoch_loss = 0
    epoch_acc = 0
    epoch_f1 = 0
    model.train()
    for batch in iterator:
        optimizer.zero_grad()
        predictions = model(batch[0].to(device))
        loss = criterion(predictions, batch[1].to(device, dtype=torch.float).view(-1, 1))
        acc = binary_accuracy(predictions, batch[1].to(device, dtype=torch.float).view(-1, 1))
        f1 = f1_score(batch[1].to(device, dtype=torch.float).view(-1, 1).detach().cpu().numpy(), predictions.detach().cpu().numpy() > 0.5)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
        epoch_acc += acc.item()
        epoch_f1 += f1
    return epoch_loss / len(iterator), epoch_acc / len(iterator), epoch_f1 / len(iterator)

# Define the evaluation function
def evaluate(model, iterator, criterion):
    epoch_loss = 0
    epoch_acc = 0
    epoch_f1 = 0
    model.eval()
    with torch.no_grad():
        for batch in iterator:
            predictions = model(batch[0].to(device))
            loss = criterion(predictions, batch[1].to(device, dtype=torch.float).view(-1, 1))
            acc = binary_accuracy(predictions, batch[1].to(device, dtype=torch.float).view(-1, 1))
            f1 = f1_score(batch[1].to(device, dtype=torch.float).view(-1, 1).cpu().numpy(), predictions.cpu().numpy() > 0.5)
            epoch_loss += loss.item()
            epoch_acc += acc.item()
            epoch_f1 += f1
    return epoch_loss / len(iterator), epoch_acc / len(iterator), epoch_f1 / len(iterator)

# Define the accuracy function
def binary_accuracy(preds, y):
    rounded_preds = torch.round(torch.sigmoid(preds))
    correct = (rounded_preds == y).float()
    acc = correct.sum() / len(correct)
    return acc
```

```python
# Define the training function
def train(model, iterator, optimizer, criterion):
    epoch_loss = 0
    epoch_acc = 0
    model.train()
    for batch in iterator:
        optimizer.zero_grad()
        predictions = model(batch[0].to(device))
        loss = criterion(predictions, batch[1].to(device, dtype=torch.float).view(-1, 1))
        acc = binary_accuracy(predictions, batch[1].to(device, dtype=torch.float).view(-1, 1))
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
        epoch_acc += acc.item()
    return epoch_loss / len(iterator), epoch_acc / len(iterator)

# Define the evaluation function
def evaluate(model, iterator, criterion):
    epoch_loss = 0
    epoch_acc = 0
    model.eval()
    with torch.no_grad():
        for batch in iterator:
            predictions = model(batch[0].to(device))
            loss = criterion(predictions, batch[1].to(device, dtype=torch.float).view(-1, 1))
            acc = binary_accuracy(predictions, batch[1].to(device, dtype=torch.float).view(-1, 1))
            epoch_loss += loss.item()
            epoch_acc += acc.item()
    return epoch_loss / len(iterator), epoch_acc / len(iterator)
```

```python
# Repeat 5 times with different random seeds and take the average
test_accs = []
test_f1s = []
test_roc_aucs = []
test_dep_f1s = []
for i in range(5):
    # Define the hyperparameters
    vocab_size = train_dataset.get_vocab_size()
    embedding_dim = 100
    num_filters = 100
    filter_sizes = [2, 3, 4]
    output_dim = 1
    dropout = 0.5
    epochs = 50
    
    # Initialize the model
    model = CharCNN(vocab_size, embedding_dim, num_filters, filter_sizes, output_dim, dropout)
    
    # Define the optimizer and loss function
    optimizer = optim.Adam(model.parameters())
    criterion = nn.BCEWithLogitsLoss()
    
    # Define the device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    
    # Move the model to the device
    model = model.to(device)
    criterion = criterion.to(device)
    
    # Train the model
    for epoch in tqdm(range(epochs)):
        train_loss, train_acc = train(model, train_loader, optimizer, criterion)
        test_loss, test_acc = evaluate(model, test_loader, criterion)
    
    # Evaluate the model
    test_loss, test_acc = evaluate(model, test_loader, criterion)
    print(f'Test Loss: {test_loss:.3f} | Test Acc: {test_acc*100:.2f}%')
    
    # Precision, Recall, F1-score
    y_pred = []
    y_true = []
    model.eval()
    with torch.no_grad():
        for batch in test_loader:
            predictions = model(batch[0].to(device)).squeeze(1)
            rounded_preds = torch.round(torch.sigmoid(predictions))
            y_pred.extend(rounded_preds.tolist())
            y_true.extend(batch[1].tolist())
    
    # Calculate the metrics
    print(f"Accuracy: {accuracy_score(y_true, y_pred)}")
    print(f"Precision: {precision_score(y_true, y_pred)}")
    print(f"Recall: {recall_score(y_true, y_pred)}")
    print(f"F1-score: {f1_score(y_true, y_pred)}")

    # Save the metrics
    test_accs.append(test_acc)
    test_f1s.append(f1_score(y_true, y_pred))
    test_roc_aucs.append(roc_auc_score(y_true, y_pred))
    test_dep_f1s.append(f1_score(np.array(y_true)[dep_test_data_index], np.array(y_pred)[dep_test_data_index]))
```

```python
# Use sentence-BERT to classify the sentences
model = SentenceTransformer('all-mpnet-base-v2')

# Get the embeddings
train_embeddings = model.encode(train_data['text'].tolist())
test_embeddings = model.encode(test_data['text'].tolist())

# Convert the embeddings to tensors
train_embeddings = torch.tensor(train_embeddings)
test_embeddings = torch.tensor(test_embeddings)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
import torch.optim as optim
```

```python

# Convert the labels to tensors
train_labels = torch.tensor((train_data['offensive'] == 'Y').astype(int).tolist())
test_labels = torch.tensor((test_data['offensive'] == 'Y').astype(int).tolist())

# Create the dataset
train_dataset = TensorDataset(train_embeddings, train_labels)
test_dataset = TensorDataset(test_embeddings, test_labels)

# Create the dataloaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Define the model
class EmbeddingClassifier(nn.Module):
    def __init__(self, freeze_bert=False):
        super(EmbeddingClassifier, self).__init__()
        self.fc = nn.Linear(768, 1)
        self.dropout = nn.Dropout(0.3)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        x = self.fc(x)
        x = self.dropout(x)
        x = self.sigmoid(x)
        return x

# Instantiate the model
model = EmbeddingClassifier(freeze_bert=True)
model = model.to(device)

# Define the optimizer and loss function
optimizer = optim.Adam(model.parameters())
```

```python
criterion = nn.BCEWithLogitsLoss()
```

```python
# Train the model. repeat 5 times with different seeds
test_accs = []
test_f1s = []
test_rocs = []
test_dep_f1s = []
for i in range(5):
    torch.manual_seed(i)
    epochs = 100
    for epoch in tqdm(range(epochs)):
        model.train()
        for batch in train_loader:
            optimizer.zero_grad()
            predictions = model(batch[0].to(device)).squeeze(1)
            loss = criterion(predictions, batch[1].to(device, dtype=torch.float))
            loss.backward()
            optimizer.step()
        # Evaluate the model
        model.eval()
        test_loss = 0
        for batch in test_loader:
            predictions = model(batch[0].to(device)).squeeze(1)
            loss = criterion(predictions, batch[1].to(device, dtype=torch.float))
            test_loss += loss.item()

    # Evaluate the model
    model.eval()
    y_pred = []
    y_true = []
    for batch in test_loader:
        predictions = model(batch[0].to(device)).squeeze(1)
        y_pred.extend((predictions > 0.5).cpu().numpy())
        y_true.extend(batch[1].cpu().numpy())

    # Print the evaluation metrics
    print(f"Accuracy: {accuracy_score(y_true, y_pred)}")
    print(f"Precision: {precision_score(y_true, y_pred)}")
    print(f"Recall: {recall_score(y_true, y_pred)}")
    print(f"F1-score: {f1_score(y_true, y_pred)}")

    print("Evaluation metrics only on those context_dependent == 'Y'")
    indices = [i for i, x in enumerate(test_data.context_dependent) if x == 'Y']
    print(f"True : False = {sum(np.array(y_true)[indices])} : {len(np.array(y_true)[indices]) - sum(np.array(y_true)[indices])}")
    print(f"Accuracy: {accuracy_score(np.array(y_true)[indices], np.array(y_pred)[indices])}")
    print(f"Precision: {precision_score(np.array(y_true)[indices], np.array(y_pred)[indices])}")
    print(f"Recall: {recall_score(np.array(y_true)[indices], np.array(y_pred)[indices])}")
    print(f"F1-score: {f1_score(np.array(y_true)[indices], np.array(y_pred)[indices])}")
    test_accs.append(accuracy_score(y_true, y_pred))
    test_rocs.append(roc_auc_score(y_true, y_pred))
    test_f1s.append(f1_score(y_true, y_pred))
    test_dep_f1s.append(f1_score(np.array(y_true)[dep_test_data_index], np.array(y_pred)[dep_test_data_index]))

print(f"Average accuracy: {np.mean(test_accs)}")
print(f"Average roc_auc_score: {np.mean(test_rocs)}")
print(f"Average f1-score: {np.mean(test_f1s)}")
print(f"Average f1-score on context dependent: {np.mean(test_dep_f1s)}")
```
